```python
# ===========================
#   IMPORTS & CONFIG
# ===========================
print("üîÑ Importing libraries...")
import os
import cv2
import random
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from concurrent.futures import ThreadPoolExecutor
import gc

warnings.filterwarnings("ignore")
sns.set(style="whitegrid")

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications import EfficientNetB5
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, Input
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, Callback
from tensorflow.keras.optimizers import Adam

from sklearn.metrics import confusion_matrix, classification_report, cohen_kappa_score, accuracy_score
from sklearn.model_selection import train_test_split

print("‚úî Imports complete.\n")

# ===========================
#   GPU OPTIMIZATION
# ===========================
print("‚ö° Configuring GPU for optimal performance...")
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"‚úî Found {len(gpus)} GPU(s)")
    except RuntimeError as e:
        print(f"‚ö† GPU config error: {e}")
else:
    print("‚ö† No GPU found - training will be slow!")

# Enable XLA compilation for faster ops
tf.config.optimizer.set_jit(True)
print("‚úî GPU optimization complete.\n")

# ===========================
#   SEEDING
# ===========================
def seed_everything(seed=42):
    print(f"üîí Setting seed: {seed}")
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

seed_everything(42)

# ===========================
#   PARAMS - OPTIMIZED
# ===========================
HEIGHT = 320
WIDTH = 320
CHANNELS = 3
BATCH_SIZE = 16  # INCREASED from 8
EPOCHS = 40
WARMUP_EPOCHS = 5
LR = 1e-4
LR_WARMUP = 1e-3
CACHE_DIR = 'preprocessed_cache'

print("üìè Image size:", HEIGHT, "x", WIDTH)
print("üì¶ Batch size:", BATCH_SIZE)
print("üéØ Warmup epochs:", WARMUP_EPOCHS)
print("üéØ Full training epochs:", EPOCHS)
print("üéØ Total epochs:", WARMUP_EPOCHS + EPOCHS, "\n")

# ===========================
#   LOAD DATA
# ===========================
print("üìÇ Loading APTOS 2019 dataset...")

df = pd.read_csv("/kaggle/input/aptos2019-blindness-detection/train.csv")
df.columns = ["id_code", "diagnosis"]
df["image_path"] = "/kaggle/input/aptos2019-blindness-detection/train_images/" + df["id_code"] + ".png"

print(f"‚úî Total images: {len(df)}")
print(f"\n‚úî Class distribution:")
for i in range(5):
    count = (df['diagnosis'] == i).sum()
    pct = count / len(df) * 100
    print(f"   Class {i}: {count:4d} images ({pct:5.2f}%)")
print()

# ===========================
#   PREPROCESSING FUNCTIONS
# ===========================
print("üîß Setting up preprocessing pipeline...")

def crop_image(img, tol=7):
    """Remove black borders from fundus image"""
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    mask = gray > tol
    if not mask.any():
        return img
    return img[np.ix_(mask.any(1), mask.any(0))]

def circle_crop(img):
    """Apply circular crop to fundus image"""
    img = crop_image(img)
    h, w, _ = img.shape
    side = max(h, w)
    img = cv2.resize(img, (side, side))
    
    x, y = side // 2, side // 2
    r = min(x, y)
    
    mask = np.zeros((side, side), np.uint8)
    cv2.circle(mask, (x, y), r, 1, -1)
    img = cv2.bitwise_and(img, img, mask=mask)
    return crop_image(img)

def preprocess_image(img_path):
    """
    Full preprocessing pipeline:
    1. Load image
    2. Circle crop
    3. Resize
    4. Ben Graham's preprocessing (contrast enhancement)
    """
    if not os.path.exists(img_path):
        print(f"‚ùå Missing image: {img_path}")
        return np.zeros((HEIGHT, WIDTH, 3), dtype=np.uint8)
    
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = circle_crop(image)
    image = cv2.resize(image, (HEIGHT, WIDTH))
    # Ben Graham's preprocessing - subtract local average color
    image = cv2.addWeighted(image, 4, cv2.GaussianBlur(image, (0, 0), 10), -4, 128)
    return image

print("‚úî Preprocessing functions ready.\n")

# ===========================
#   CACHE PREPROCESSED IMAGES
# ===========================
print("üíæ Preprocessing and caching images to disk...")
print("   This will take 2-3 minutes but saves hours during training!\n")

def cache_preprocessed_images(df, cache_dir=CACHE_DIR):
    """
    Preprocess ALL images once and cache to disk
    KEY OPTIMIZATION: No more on-the-fly preprocessing!
    """
    if not os.path.exists(cache_dir):
        os.makedirs(cache_dir)
    
    cached_count = 0
    new_count = 0
    
    def process_single(idx_row):
        nonlocal cached_count, new_count
        idx, row = idx_row
        cache_path = os.path.join(cache_dir, f"{row['id_code']}.npy")
        
        if os.path.exists(cache_path):
            cached_count += 1
            return cache_path
        
        # Preprocess and save
        img = preprocess_image(row['image_path'])
        np.save(cache_path, img)
        new_count += 1
        
        if (idx + 1) % 500 == 0:
            print(f"   Processed {idx + 1}/{len(df)} images...")
        
        return cache_path
    
    # Use ThreadPoolExecutor for parallel preprocessing
    with ThreadPoolExecutor(max_workers=4) as executor:
        cache_paths = list(executor.map(process_single, df.iterrows()))
    
    df['cache_path'] = cache_paths
    
    if new_count > 0:
        print(f"‚úî Preprocessed {new_count} new images")
    if cached_count > 0:
        print(f"‚úî Loaded {cached_count} cached images")
    print(f"‚úî Total: {len(df)} images ready in {cache_dir}\n")
    
    return df

# Cache all images
df = cache_preprocessed_images(df)

# ===========================
#   TRAIN/VAL SPLIT
# ===========================
print("‚úÇ Creating stratified train/validation split (80/20)...")

X_train, X_val = train_test_split(
    df, 
    test_size=0.2, 
    random_state=42, 
    stratify=df["diagnosis"]
)

X_train = X_train.reset_index(drop=True)
X_val = X_val.reset_index(drop=True)

print(f"\n‚úî Train size: {len(X_train)} images ({len(X_train)/len(df)*100:.1f}%)")
print(f"‚úî Validation size: {len(X_val)} images ({len(X_val)/len(df)*100:.1f}%)")

print(f"\n‚úî Training set class distribution:")
for i in range(5):
    count = (X_train['diagnosis'] == i).sum()
    print(f"   Class {i}: {count:4d} images")

print(f"\n‚úî Validation set class distribution:")
for i in range(5):
    count = (X_val['diagnosis'] == i).sum()
    print(f"   Class {i}: {count:3d} images")
print()

# ==== SHOW 5 SAMPLES OF PREPROCESSED IMAGES ====
import random
plt.figure(figsize=(15, 15))

sample_idx = random.sample(range(len(X_train)), 25)

for i, idx in enumerate(sample_idx):
    row = X_train.iloc[idx]
    img = np.load(row['cache_path'])
    label = row['diagnosis']

    plt.subplot(5, 5, i+1)
    plt.imshow(img.astype("uint8"))
    plt.title(f"Label: {int(label)}")
    plt.axis("off")

plt.suptitle("Sample Preprocessed Images", fontsize=16)
plt.show()


# ===========================
#   FAST TF.DATA PIPELINE
# ===========================
print("‚ö° Building optimized tf.data pipeline...")

def make_dataset(df, batch_size=BATCH_SIZE, augment=False, shuffle=True):
    """
    Fast tf.data pipeline - loads from cached numpy files
    Much faster than Python generators!
    """
    cache_paths = df['cache_path'].values
    labels = df['diagnosis'].values.astype(np.float32)  # Float for regression
    
    dataset = tf.data.Dataset.from_tensor_slices((cache_paths, labels))
    
    def load_cached_image(path, label):
        """Load preprocessed image from cache"""
        def _load(path_bytes):
            img = np.load(path_bytes.decode('utf-8'))
            return img.astype(np.float32) / 255.0
        
        img = tf.numpy_function(_load, [path], tf.float32)
        img.set_shape([HEIGHT, WIDTH, 3])
        
        if augment:
            # Fast augmentation using TF ops
            img = tf.image.random_flip_left_right(img)
            img = tf.image.random_flip_up_down(img)
            img = tf.image.random_brightness(img, 0.2)
            img = tf.image.random_contrast(img, 0.8, 1.2)
            
            # Random rotation (0, 90, 180, 270 degrees)
            if tf.random.uniform(()) > 0.5:
                k = tf.random.uniform((), maxval=4, dtype=tf.int32)
                img = tf.image.rot90(img, k=k)
        
        return img, label
    
    dataset = dataset.map(load_cached_image, num_parallel_calls=tf.data.AUTOTUNE)
    
    if shuffle:
        dataset = dataset.shuffle(buffer_size=1000, seed=42)
    
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    
    return dataset

train_ds = make_dataset(X_train, augment=True, shuffle=True)
val_ds = make_dataset(X_val, augment=False, shuffle=False)

print("‚úî Fast data pipeline ready.")
print(f"‚úî Training batches per epoch: {len(train_ds)}")
print(f"‚úî Validation batches per epoch: {len(val_ds)}\n")

# ===========================
#   CUSTOM METRICS
# ===========================
import tensorflow as tf

class RegressionAccuracy(tf.keras.metrics.Metric):
    def __init__(self, name='reg_accuracy', **kwargs):
        super().__init__(name=name, **kwargs)
        self.total = self.add_weight(name='total', initializer='zeros')
        self.count = self.add_weight(name='count', initializer='zeros')

    def update_state(self, y_true, y_pred, sample_weight=None):
        y_true = tf.cast(tf.reshape(y_true, [-1]), tf.float32)
        y_pred = tf.round(tf.clip_by_value(tf.reshape(y_pred, [-1]), 0, 4))
        matches = tf.cast(tf.equal(y_true, y_pred), tf.float32)
        batch_acc = tf.reduce_mean(matches)
        self.total.assign_add(batch_acc)
        self.count.assign_add(1.0)

    def result(self):
        return tf.math.divide_no_nan(self.total, self.count)

    def reset_state(self):
        self.total.assign(0)
        self.count.assign(0)


# ===========================
#   CUSTOM CALLBACK FOR KAPPA
# ===========================
class KappaCallback(Callback):
    """
    Calculate Quadratic Kappa at end of each epoch
    """
    def __init__(self, validation_data, X_val_df):
        super().__init__()
        self.validation_data = validation_data
        self.X_val_df = X_val_df
        self.kappa_history = []
    
    def on_epoch_end(self, epoch, logs=None):
        # Get predictions
        val_preds = self.model.predict(self.validation_data, verbose=0)
        val_preds_rounded = np.clip(np.round(val_preds.flatten()), 0, 4).astype(int)
        true_labels = self.X_val_df['diagnosis'].values
        
        # Calculate kappa
        kappa = cohen_kappa_score(true_labels, val_preds_rounded, weights='quadratic')
        self.kappa_history.append(kappa)
        
        logs['val_kappa'] = kappa
        print(f"\n   Validation Kappa: {kappa:.4f}")

# ===========================
#   MODEL BUILDING
# ===========================
print("üèó Building EfficientNet-B5 model...")

def build_model():
    """
    Regression model for DR severity prediction
    Output: continuous value 0-4, rounded to nearest integer for evaluation
    """
    inp = Input(shape=(HEIGHT, WIDTH, 3), name='input_image')
    
    # EfficientNetB5 backbone
    base = EfficientNetB5(
        include_top=False, 
        weights="imagenet", 
        input_tensor=inp
    )
    
    x = GlobalAveragePooling2D(name='global_pool')(base.output)
    x = Dropout(0.5, name='dropout_1')(x)
    x = Dense(2048, activation="relu", name='dense_1')(x)
    x = Dropout(0.5, name='dropout_2')(x)
    
    # Regression output (0-4 continuous)
    out = Dense(1, activation="linear", name='output')(x)
    
    model = Model(inp, out, name='DR_EfficientNetB5_Regression')
    
    return model

model = build_model()

print(f"\n‚úî Model architecture:")
print(f"   Model name: {model.name}")
print(f"   Input shape: {model.input_shape}")
print(f"   Output shape: {model.output_shape}")
print(f"   Total parameters: {model.count_params():,}")
print(f"   Total layers: {len(model.layers)}")
print()

# ===========================
#   PHASE 1: WARMUP TRAINING
# ===========================
print("="*60)
print("üî• PHASE 1: WARMUP TRAINING (FROZEN BACKBONE)")
print("="*60)
print(f"Training only the head layers for {WARMUP_EPOCHS} epochs")
print(f"This helps stabilize the randomly initialized head before fine-tuning\n")

# Freeze all except last 5 layers
for layer in model.layers[:-5]:
    layer.trainable = False

trainable_params = sum([tf.size(w).numpy() for w in model.trainable_weights])
print(f"‚úî Frozen backbone - Trainable parameters: {trainable_params:,}")

model.compile(
    optimizer=Adam(learning_rate=LR_WARMUP),
    loss='mse',
    metrics=['mae', RegressionAccuracy()]
)

kappa_callback_warmup = KappaCallback(val_ds, X_val)

print(f"\n{'='*60}")
print("Starting warmup training...")
print(f"{'='*60}\n")

warmup_history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=WARMUP_EPOCHS,
    callbacks=[kappa_callback_warmup],
    verbose=1
)

print(f"\n{'='*60}")
print("‚úî Warmup phase complete!")
print(f"{'='*60}\n")

# ===========================
#   PHASE 2: FULL TRAINING
# ===========================
print("="*60)
print("üöÄ PHASE 2: FULL FINE-TUNING (UNFROZEN BACKBONE)")
print("="*60)
print(f"Training all layers for up to {EPOCHS} epochs")
print(f"Early stopping will halt training if validation loss stops improving\n")

# Unfreeze all layers
for layer in model.layers:
    layer.trainable = True

trainable_params = sum([tf.size(w).numpy() for w in model.trainable_weights])
print(f"‚úî Unfrozen all layers - Trainable parameters: {trainable_params:,}")

model.compile(
    optimizer=Adam(learning_rate=LR),
    loss='mse',
    metrics=['mae', RegressionAccuracy()]
)

# Callbacks
kappa_callback_full = KappaCallback(val_ds, X_val)

callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=7,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        patience=3,
        factor=0.5,
        min_lr=1e-7,
        verbose=1
    ),
    ModelCheckpoint(
        'best_dr_model.h5',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    ),
    kappa_callback_full
]

print(f"\n{'='*60}")
print("Starting full training...")
print(f"{'='*60}\n")

full_history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

print(f"\n{'='*60}")
print("‚úî Full training complete!")
print(f"{'='*60}\n")

# ===========================
#   TRAINING CURVES
# ===========================
print("üìä Plotting training curves...\n")

# Combine warmup and full training history
all_loss = warmup_history.history['loss'] + full_history.history['loss']
all_val_loss = warmup_history.history['val_loss'] + full_history.history['val_loss']
all_acc = warmup_history.history['reg_accuracy'] + full_history.history['reg_accuracy']
all_val_acc = warmup_history.history['val_reg_accuracy'] + full_history.history['val_reg_accuracy']
all_kappa = kappa_callback_warmup.kappa_history + kappa_callback_full.kappa_history

fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Loss curve
axes[0].plot(all_loss, label='Train Loss (MSE)', linewidth=2, alpha=0.8, color='#2E86AB')
axes[0].plot(all_val_loss, label='Val Loss (MSE)', linewidth=2, alpha=0.8, color='#A23B72')
axes[0].axvline(x=WARMUP_EPOCHS-0.5, color='red', linestyle='--', alpha=0.5, linewidth=2, label='Unfreeze')
axes[0].set_xlabel('Epoch', fontsize=12, fontweight='bold')
axes[0].set_ylabel('MSE Loss', fontsize=12, fontweight='bold')
axes[0].set_title('Loss Curve', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

# Accuracy curve
axes[1].plot(all_acc, label='Train Accuracy', linewidth=2, alpha=0.8, color='#2E86AB')
axes[1].plot(all_val_acc, label='Val Accuracy', linewidth=2, alpha=0.8, color='#A23B72')
axes[1].axvline(x=WARMUP_EPOCHS-0.5, color='red', linestyle='--', alpha=0.5, linewidth=2, label='Unfreeze')
axes[1].set_xlabel('Epoch', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Accuracy (Rounded)', fontsize=12, fontweight='bold')
axes[1].set_title('Accuracy Curve', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

# Kappa curve
axes[2].plot(all_kappa, label='Validation Kappa', linewidth=2, alpha=0.8, color='#F18F01', marker='o')
axes[2].axvline(x=WARMUP_EPOCHS-0.5, color='red', linestyle='--', alpha=0.5, linewidth=2, label='Unfreeze')
axes[2].set_xlabel('Epoch', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Quadratic Kappa', fontsize=12, fontweight='bold')
axes[2].set_title('Kappa Score Curve', fontsize=14, fontweight='bold')
axes[2].legend(fontsize=10)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ===========================
#   FINAL EVALUATION
# ===========================
print("üîÆ Evaluating final model on validation set...\n")

# Get predictions
print("Generating predictions...")
preds = model.predict(val_ds, verbose=1)
true_labels = X_val["diagnosis"].values

# Round predictions to nearest integer
val_preds = np.clip(np.round(preds.flatten()), 0, 4).astype(int)

# Calculate metrics
kappa = cohen_kappa_score(true_labels, val_preds, weights="quadratic")
accuracy = accuracy_score(true_labels, val_preds)

print(f"\n{'='*60}")
print(f"üìä FINAL VALIDATION RESULTS")
print(f"{'='*60}")
print(f"‚úî Validation Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"‚úî Quadratic Kappa Score: {kappa:.4f}")
print(f"‚úî Best Kappa during training: {max(all_kappa):.4f}")
print(f"{'='*60}\n")

# ===========================
#   CONFUSION MATRIX
# ===========================
print("üìà Generating confusion matrices...\n")

cm = confusion_matrix(true_labels, val_preds)
cm_norm = cm.astype('float') / cm.sum(axis=1, keepdims=True)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Raw counts
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=axes[0], 
            cbar_kws={'label': 'Count'}, linewidths=0.5, linecolor='gray')
axes[0].set_xlabel('Predicted Class', fontsize=12, fontweight='bold')
axes[0].set_ylabel('True Class', fontsize=12, fontweight='bold')
axes[0].set_title('Confusion Matrix (Raw Counts)', fontsize=14, fontweight='bold')

# Normalized
sns.heatmap(cm_norm, annot=True, fmt=".1%", cmap="Greens", ax=axes[1], 
            cbar_kws={'label': 'Percentage'}, linewidths=0.5, linecolor='gray')
axes[1].set_xlabel('Predicted Class', fontsize=12, fontweight='bold')
axes[1].set_ylabel('True Class', fontsize=12, fontweight='bold')
axes[1].set_title('Confusion Matrix (Normalized by True Class)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# ===========================
#   CLASSIFICATION REPORT
# ===========================
print("\nüìÑ Detailed Classification Report:\n")
print(classification_report(
    true_labels, 
    val_preds,
    target_names=['No DR (0)', 'Mild (1)', 'Moderate (2)', 'Severe (3)', 'Proliferative (4)'],
    digits=4
))

# Per-class detailed analysis
print("\nüìä Per-Class Detailed Analysis:\n")
print(f"{'Class':<6} {'Precision':<12} {'Recall':<12} {'F1-Score':<12} {'Support':<10}")
print("-" * 60)

for i in range(5):
    class_mask = true_labels == i
    if class_mask.sum() > 0:
        tp = ((val_preds == i) & (true_labels == i)).sum()
        fp = ((val_preds == i) & (true_labels != i)).sum()
        fn = ((val_preds != i) & (true_labels == i)).sum()
        
        precision = tp / max((tp + fp), 1)
        recall = tp / max((tp + fn), 1)
        f1 = 2 * (precision * recall) / max((precision + recall), 1e-10)
        support = class_mask.sum()
        
        severity = ['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative'][i]
        print(f"{i} ({severity:<15}) {precision:>6.3f}        {recall:>6.3f}        {f1:>6.3f}        {support:>4d}")

# ===========================
#   PREDICTION DISTRIBUTION
# ===========================
print("\nüìä Prediction distribution analysis:\n")

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# True vs Predicted distribution
classes = ['No DR\n(0)', 'Mild\n(1)', 'Moderate\n(2)', 'Severe\n(3)', 'Prolif.\n(4)']
x = np.arange(5)
width = 0.35

true_counts = [np.sum(true_labels == i) for i in range(5)]
pred_counts = [np.sum(val_preds == i) for i in range(5)]

axes[0].bar(x - width/2, true_counts, width, label='True Labels', alpha=0.8, color='#2E86AB')
axes[0].bar(x + width/2, pred_counts, width, label='Predictions', alpha=0.8, color='#A23B72')
axes[0].set_xlabel('DR Severity Class', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Count', fontsize=12, fontweight='bold')
axes[0].set_title('True vs Predicted Distribution', fontsize=14, fontweight='bold')
axes[0].set_xticks(x)
axes[0].set_xticklabels(classes)
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)

# Prediction errors
errors = val_preds - true_labels
axes[1].hist(errors, bins=np.arange(-4.5, 5.5, 1), alpha=0.7, color='#F18F01', edgecolor='black')
axes[1].axvline(x=0, color='red', linestyle='--', linewidth=2, label='Perfect Prediction')
axes[1].set_xlabel('Prediction Error (Predicted - True)', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Frequency', fontsize=12, fontweight='bold')
axes[1].set_title('Prediction Error Distribution', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# Calculate mean absolute error
mae = np.mean(np.abs(errors))
print(f"Mean Absolute Error: {mae:.4f}")
print(f"Perfect predictions: {np.sum(errors == 0)} / {len(errors)} ({np.sum(errors == 0)/len(errors)*100:.2f}%)")
print(f"Off by 1: {np.sum(np.abs(errors) == 1)} ({np.sum(np.abs(errors) == 1)/len(errors)*100:.2f}%)")
print(f"Off by 2+: {np.sum(np.abs(errors) >= 2)} ({np.sum(np.abs(errors) >= 2)/len(errors)*100:.2f}%)")

# ===========================
#   SAVE FINAL MODEL
# ===========================
print("\n" + "="*60)
print("üíæ SAVING MODEL FOR DEPLOYMENT")
print("="*60)

# Save final model
model.save("dr_model.h5")
print("‚úî Final model saved as 'dr_model.h5'")
```
